```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('6 class csv.csv')
df.head()
```

```python
print(df.info())
print("--------")
print(df.columns)
print("--------")

print(list(set(df['Star color'])))
```

```python
blue = ['Blue', 'Blue white ','Blue-white','Blue-White','Blue white','Blue White','Blue ']
orange = ['Orange','Orange-Red','Red','Pale yellow orange']
yellow = ['Yellowish', 'yellow-white', 'yellowish','Yellowish White','White-Yellow']
white = ['White', 'white','Whitish']



blueStar = df[df["Star color"].isin(blue)]
orangeStar = df[df["Star color"].isin(orange)]
yellowStar = df[df["Star color"].isin(yellow)]
whiteStar = df[df["Star color"].isin(white)]
blueStar
```

```python
plt.scatter(blueStar['Temperature (K)'],blueStar['Luminosity(L/Lo)'],c='blue',s=2,label='blueStar')
plt.scatter(orangeStar['Temperature (K)'],orangeStar['Luminosity(L/Lo)'],c='orange',s=2,label='orangeStar')
plt.scatter(yellowStar['Temperature (K)'],yellowStar['Luminosity(L/Lo)'],c='yellow',s=2,label='yellowStar')
plt.scatter(whiteStar['Temperature (K)'],whiteStar['Luminosity(L/Lo)'],c='red',s=2,label='whiteStar')

```

```python
temp = df
indexOutli = temp[temp['Temperature (K)']>20000].index
temp.drop(indexOutli,inplace=True)
indexOutli = temp[temp['Luminosity(L/Lo)']>400000].index
temp.drop(indexOutli,inplace=True)
```

```python
blueStar = temp[temp["Star color"].isin(blue)]
orangeStar = temp[temp["Star color"].isin(orange)]
yellowStar = temp[temp["Star color"].isin(yellow)]
whiteStar = temp[temp["Star color"].isin(white)]
plt.scatter(blueStar['Temperature (K)'],blueStar['Luminosity(L/Lo)'],c='blue',s=2,label='blueStar')
plt.scatter(orangeStar['Temperature (K)'],orangeStar['Luminosity(L/Lo)'],c='orange',s=2,label='orangeStar')
plt.scatter(yellowStar['Temperature (K)'],yellowStar['Luminosity(L/Lo)'],c='yellow',s=2,label='yellowStar')
plt.scatter(whiteStar['Temperature (K)'],whiteStar['Luminosity(L/Lo)'],c='red',s=2,label='whiteStar')

plt.legend()
plt.show()
```
